## MLP on Banana 

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from datetime import datetime
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

# Load data from file
data = np.loadtxt('banana.dat', delimiter=',')

# Split data into input and target arrays
X = data[:, :2]
y = data[:, 2]

# Split data into training and test sets with 70% for training and 30% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize MLP model
model = Sequential()
model.add(Dense(1024, activation='relu', input_shape=(2,)))
model.add(Dense(512, activation='relu'))
model.add(Dense(1))
model.summary()

# Define optimizer with learning rate
opt = Adam(learning_rate=0.005)
# Compile the model
model.compile(optimizer=opt, loss='mean_squared_error')

# Train the model
start_time = datetime.now()
model.fit(X_train, y_train, epochs=100)
end_time = datetime.now()
train_time = end_time - start_time

# Evaluate the model accuracy using test data
num_tests = 30
acc_sum = 0
start_time = datetime.now()
for _ in range(num_tests):
    #test_data = np.random.rand(10, 2)  # Generate random test data
    y_pred = model.predict(X_test)
    acc = np.mean(np.abs(y_pred - 1) < 2)  # Calculate accuracy
    acc_sum += acc
end_time = datetime.now()
test_time = end_time - start_time

# Calculate final accuracy as the average of all tests
final_acc = acc_sum / num_tests

# Parameters for print results
lrb = 0.005
trtb = train_time
tetb =  test_time
accb = final_acc
nib = 1024
nhb = 512
nob = 1

# Print final accuracy
print(f'Final accuracy: {final_acc}')
print(f'Training time: {train_time.total_seconds()} seconds')
print(f'Testing time: {test_time.total_seconds()} seconds')

# Plot the input data points with different colors based on their class labels
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('Input Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Define a mesh grid to plot the decision boundary
x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary learned by the model
plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu_r)
plt.title('Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Radial Basis Function Algorithm

## RBF on Flame

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances
from scipy.linalg import pinv
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from datetime import datetime

# Load data from file
data = np.loadtxt('Flame.txt', delimiter=',')

# Split data into input and target arrays
X = data[:, :2]
y = data[:, 2].reshape(-1, 1)

# Split data into train and test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Perform K-means clustering on the training set to determine RBF centers
n_centers = 10  # Number of RBF centers
start_time = datetime.now()
kmeans = KMeans(n_clusters=n_centers, n_init=10, random_state=42)
kmeans.fit(X_train)
centers = kmeans.cluster_centers_
end_time = datetime.now()
train_time = end_time - start_time

# Calculate RBF widths based on pairwise distances between centers
widths = np.mean(pairwise_distances(X_train, centers, metric='euclidean'))

# Compute RBF activations for both train and test sets
rbf_activations_train = np.exp(-pairwise_distances(X_train, centers, metric='euclidean') ** 2 / (2 * widths ** 2))
rbf_activations_test = np.exp(-pairwise_distances(X_test, centers, metric='euclidean') ** 2 / (2 * widths ** 2))

# Add bias term to RBF activations for both train and test sets
rbf_activations_train = np.hstack((rbf_activations_train, np.ones((X_train.shape[0], 1))))
rbf_activations_test = np.hstack((rbf_activations_test, np.ones((X_test.shape[0], 1))))

# One-hot encode the target variable for both train and test sets
enc = OneHotEncoder()
y_train_onehot = enc.fit_transform(y_train).toarray()
y_test_onehot = enc.transform(y_test).toarray()

# Train and test the RBF model for 30 iterations
num_iterations = 30
accuracies = []

start_time = datetime.now()
for _ in range(num_iterations):
    # Compute RBF model weights using Moore-Penrose pseudoinverse on the training set
    weights = np.dot(pinv(rbf_activations_train), y_train_onehot)

    # Predict the class labels for the test set
    y_pred = np.argmax(np.dot(rbf_activations_test, weights), axis=1) + 1

    # Calculate test accuracy for this iteration
    accuracy = np.mean(y_pred == y_test.flatten()) 
    accuracies.append(accuracy)
end_time = datetime.now()
test_time = end_time - start_time

# Calculate average accuracy over all iterations
final_acc = np.mean(accuracies)
print(f'Final accuracy: {final_acc}')
print(f'Training time: {train_time.total_seconds()} seconds')
print(f'Testing time: {test_time.total_seconds()} seconds')

# Parameters for print results
ncrf = n_centers
trtrf = train_time
tetrf =  test_time
accrf = final_acc


# Plot the input data points with different colors based on their class labels
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('Input Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Probability Neural Network (PNN) Algorithm

## PNN on Flame

```python
import numpy as np
from sklearn.model_selection import train_test_split
from datetime import datetime
import matplotlib.pyplot as plt

# Load data from file
data = np.loadtxt('Flame.txt', delimiter=',')

# Split data into input and target arrays
X = data[:, :2]
y = data[:, 2]

# Define the PNN function
def pnn(X_train, y_train, X_test, sigma):

    # Calculate the number of classes
    classes = np.unique(y_train)
    num_classes = len(classes)
    
    # Calculate the number of training samples
    num_train_samples = X_train.shape[0]
    
    # Initialize the output matrix
    y_pred = np.zeros((X_test.shape[0], num_classes))
    
    # Iterate over the test samples
    for i in range(X_test.shape[0]):
        # Calculate the distances between the test sample and all training samples
        distances = np.sum((X_train - X_test[i])**2, axis=1)
 
        # Apply the Gaussian kernel to the distances
        kernel = np.exp(-distances / (2 * sigma**2))
        
        # Calculate the class-wise probabilities
        class_probs = np.zeros(num_classes)
        for j in range(num_classes):
            class_idx = np.where(y_train == classes[j])[0]
            class_kernel = kernel[class_idx]
            class_probs[j] = np.sum(class_kernel) / num_train_samples
        
        # Assign the class probabilities to the output matrix
        y_pred[i] = class_probs   
    return y_pred

# Set the number of iterations to perform
num_iterations_train = 100
num_iterations_test = 30

# Set the value of sigma (the kernel bandwidth)
sigma = 1.0

# Initialize the accuracy list
accuracy_list = []

# Split data into train and test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Iterate over the number of iterations
start_time = datetime.now()
for i in range(num_iterations_train):
    # Split the data into training and test sets
    indices = np.random.permutation(X.shape[0])
    train_indices, test_indices = indices[:int(0.7*X.shape[0])], indices[int(0.7*X.shape[0]):]
    
    # Train the PNN model
    y_pred_train = pnn(X_train, y_train, X_train, sigma)
end_time = datetime.now()
train_time = end_time - start_time

start_time = datetime.now()
for i in range(num_iterations_test):
    # Make predictions on the test data
    y_pred_test = pnn(X_train, y_train, X_test, sigma)
    
    # Calculate the predicted classes for test data
    y_pred_classes = np.argmax(y_pred_test, axis=1) + 1
    
    # Calculate the accuracy for this iteration
    accuracy = np.mean(y_pred_classes == y_test)
    
    # Append accuracy to the list
    accuracy_list.append(accuracy)
end_time = datetime.now()
test_time = end_time - start_time

# Calculate the average accuracy over all iterations
average_accuracy = np.mean(accuracy_list)

# Calculate average accuracy over all iterations
print(f'Final accuracy: {average_accuracy}')
print(f'Training time: {train_time.total_seconds()} seconds')
print(f'Testing time: {test_time.total_seconds()} seconds')

# Parameters for print results
trtpf = train_time
tetpf =  test_time
accpf = average_accuracy
sigpf = sigma

# Plot the input data points with different colors based on their class labels
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title('Input Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
